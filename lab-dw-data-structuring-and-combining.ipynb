```python
import pandas as pd
```

```python
# Check all datasets
df_insurance1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
display(df_insurance1.info())
df_insurance2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
display(df_insurance2.info())
df_insurance3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
display(df_insurance3.info())
                        
```

```python
def cleaning_df(df):
    """
    Dockstring Placeholder
    
    """
    
    # Rename column names
    df = df.rename(columns = {"ST":"state"})
    df.columns = df.columns.str.lower()
    df.columns = [col.replace(" ", "_") for col in df.columns]

    # Sanitize column "gender"
    df["gender"] = df["gender"].apply(lambda x: "M" if isinstance(x, str) and x.lower().startswith("m") else ("F" if isinstance(x, str) else x))

    # Rename value names in column "states" and "education"
    df["state"] = df["state"].replace({"WA":"Washington", "AZ":"Arizona", "Cali":"California"})
    df["education"] = df["education"].replace({"Bachelors":"Bachelor"})

    # Clean values in column "customer_lifetime_value"
    try:
        df["customer_lifetime_value"] = df["customer_lifetime_value"].str.strip('%').astype(float)
    except AttributeError:
        df["customer_lifetime_value"] = df["customer_lifetime_value"].astype(float)
    
    # Cluster values in column "vehicle_class"
    df["vehicle_class"] = df["vehicle_class"].replace({"Sports Car":"Luxury", "Luxury SUV":"Luxury", "Luxury Car":"Luxury"})

    # Change data type of column "customer_lifetime_value"
    #df["customer_lifetime_value"] = df["customer_lifetime_value"].astype(float)

    # Split "number_of_open_complaints" and keep only middle value + change data type
    try:
        df["number_of_open_complaints"] = df["number_of_open_complaints"].str.split("/").str[1].astype(float)
    except AttributeError:
        df["number_of_open_complaints"] = df["number_of_open_complaints"].astype(float)

    # Drop all rows with missing values
    df = df.dropna(how="all")

    # Fill in missing values in gender with "not specified"
    df["gender"] = df["gender"].fillna("not specified")

    # Fill in missing values in customer_lifetime_value with average value
    df["customer_lifetime_value"] = df["customer_lifetime_value"].fillna(df["customer_lifetime_value"].mean())

    # Converting float columns to integer
    df_clean = df.apply(lambda col: col.astype(int) if col.dtype == 'float' else col)

    return df_clean
```

```python
clean_dataset1 = cleaning_df(df_insurance1)
print(clean_dataset1.columns)

clean_dataset2 = cleaning_df(df_insurance2)
print(clean_dataset2.columns)

clean_dataset3 = cleaning_df(df_insurance3)
print(clean_dataset3.columns)
```

```python
# Concatenate all three datasets
clean_dataset_insurance = pd.concat([clean_dataset1, clean_dataset2, clean_dataset3], axis=0).reset_index(drop=True)

# Check for missing values and for duplicates
clean_dataset_insurance.isna().sum() # no missing values 

# ??? # How are those 3 rows duplicates?
duplicates = clean_dataset_insurance.duplicated()
clean_dataset_insurance[duplicates] 

# Drop duplicates
#clean_dataset_insurance = clean_dataset_insurance.drop_duplicates()
#clean_dataset_insurance.duplicated().sum()
#display(clean_dataset_insurance)
```

The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*
